```python
import matplotlib.pyplot as plt
```

```python
timeExecutionWithIndex = 0
documentSearchedWithIndex = 1

timeExecutionWithoutIndex = 40
documentSearchedWithoutIndex = 42741
```

```python

x = ['Time Execution (With Index) - 0', 'Time Execution (Without Index) - 40']
yval = [timeExecutionWithIndex, timeExecutionWithoutIndex]
x_pos = [0,1]
plt.bar(x_pos, yval, color='red')
plt.xticks(x_pos, x, rotation="vertical")
plt.xlabel("Indexing Strategiy")
plt.ylabel("Document Searched")
plt.title("Performace difference between Index and Non-Index(Time)")
plt.show()
```

```python
x = ['Document Searched (With Index) - 1', 'DocumentSearched (Without Index) - 42741']
yval = [documentSearchedWithIndex, documentSearchedWithoutIndex]
x_pos = [0,1]
plt.bar(x_pos, yval, color='red')
plt.xticks(x_pos, x, rotation="vertical")
plt.xlabel("Indexing Strategiy")
plt.ylabel("Time Taken")
plt.title("Performace difference between Index and Non-Index(Document Search)")
plt.show()
```
