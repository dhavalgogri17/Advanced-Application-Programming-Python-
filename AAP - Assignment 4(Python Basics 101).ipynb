# Basic DataFrame Function

```python
#USING DROP DUPLICATES FUNCTIONALITY.
#THIS FUNCTION IS COMMANLY USED IN CLEANING DATA FOR DATA ANALYSIS. The below function will delete the duplicate values
#Below function will remove all the duplicates but would keep the first one.
df_dup1 = pd.DataFrame({"A":["data1", "data1", "data2", "data3", "data4", 'data1', 'data5', 'data2']})
df_dup1.drop_duplicates(subset=['A'], keep='first', inplace = True)

df_dup1
```

```python
#Below function will remove all the duplicates but would keep the last one.
df_dup2 = pd.DataFrame({"A":["data1", "data1", "data2", "data3", "data4", 'data1', 'data5', 'data2']})
df_dup2.drop_duplicates(subset=['A'], keep='last', inplace = True)

df_dup2
```

```python
#Below function will remove all the duplicates and wont keep any of those.
df_dup3 = pd.DataFrame({"A":["data1", "data1", "data2", "data3", "data4", 'data1', 'data5', 'data2']})
df_dup3.drop_duplicates(subset=['A'], keep=False, inplace = True)

df_dup3
```

# MERGE, JOIN AND CONCATENATE

```python
# Merges all the dataframes into a 1 single dataframe. Directly taken from pandas documents.
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                     index=[0, 1, 2, 3])
 

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                      index=[4, 5, 6, 7])
 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                     index=[8, 9, 10, 11])
 

frames = [df1, df2, df3]

result = pd.concat(frames)
print(result)
```

# TIME SERIES

```python
# The Time Series framework is great to work with raw data and convert it to time stamp.
# As we can look at a few example below, The date and time function takes the input as a normal english sentence 
# and converts it to a date time stamp, which can then be used for various computation like comparing dates and time.
# This can be useful to make a list of events from reading a text file or scrapping web page and analyse the data to
# find any upcoming events.


# Only Date
date1 = pd.to_datetime("14th of September, 2018")
date2 = pd.to_datetime("September 13, 2018")
date3 = pd.to_datetime("Sep 12, 2018")


# With Date and Time
date4 = pd.to_datetime("Sep 12, 2018 at 3 pm")
date5 = pd.to_datetime("09 14, 2018 at 5pm")

print(date1)
print(date2)
print(date3)
print(date4)
print(date5)
```

# STYLES

Giving a background color to a column based on the totalGoals scored in the worldcup. 

```python
data = pd.read_csv("/Users/dhavalgogri/Documents/Dhaval old computer/000 SMU study materials/Fall2018/AdvancedApplProgramming/Quest3/worldcup.csv") 
average = data["goalsScored"].mean()
print(average)
```

```python
# Color the value in the cell based on its value. If the value if high then color different
def color_goals_scored(val):
    color = 'black'
    if val < average/2:
        color = 'red'
    elif val > average/2 and val < average*1.2:
        color = 'orange'
    else:
        color = 'green'
    return 'color: %s' % color
```

```python
data1 = data.style.applymap(color_goals_scored, subset=['goalsScored'])
data1
```
